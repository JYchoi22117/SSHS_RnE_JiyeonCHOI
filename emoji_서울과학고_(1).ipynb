# ***wellness+songys***

```python
from wordcloud import WordCloud # 워드클라우드 제작 라이브러리
import pandas as pd # 데이터 프레임 라이브러리
import numpy as np # 행렬 라이브러리
import matplotlib.pyplot as plt # 워드클라우드 시각화 라이브러리
```

```python
df = pd.read_csv("/content/drive/MyDrive/연구/서울과학고/Chatbot_Data")
df
```

```python
df['A'] = df['A'].str.replace('[^가-힣]', ' ', regex = True)
df['A']
```

```python
import konlpy
kkma = konlpy.tag.Kkma() #형태소 분석기 꼬꼬마(Kkma)

nouns = df['A'].apply(kkma.nouns)
nouns
```

```python
df_word = pd.DataFrame({'word' : nouns})
df_word['count'] = df_word['word'].str.len()
df_word = df_word.query('count >= 2')
df_word
```

```python
sorted_df = df_word.sort_values(by="count", ascending=False)
print(sorted_df)
```
